```python
covid_stats = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
covid_stats.head()
```

```python
def geo_locator(row):
    try:
        point = geocode(row, provider = 'nominatim').geometry.iloc[0]
        return pd.Series({'Latitude':point.y,'Longitude':point.x, 'geometry':point})
    except:
        return None
```

```python
states = covid_stats[["sno","state_name"]]
states[['Latitude', 'Longitude', 'geometry']] = states.apply(lambda x : geo_locator(x['state_name']), axis=1)
states
```
